#3°/ Compréhension des liens des dataset et choix d'analyse

Etudier le rapport entre chaque dataset

```python
EDSData = EDSData.rename(columns={"Country Code": "CountryCode"})
EDSCountry = EDSCountry.rename(columns={"Country Code": "CountryCode"})
```

```python
datasets = {
    "EDSData": EDSData,
    "EDSFootNote": EDSFootNote,
    "EDSSeries": EDSSeries,
    "EDSCountry": EDSCountry,
    "EDSCountrySeries": EDSCountrySeries
}

for name1, df1 in datasets.items():
    for name2, df2 in datasets.items():
        if name1 != name2:
            common_columns = df1.columns.intersection(df2.columns)
            if common_columns.empty == False:
              print(f"{name1} n {name2} : {common_columns}")
```

Rien ne semble est problèmatique avec les types, par contre la colonne Unnamed : 69 ne correspond à rien

```python
EDSData_copy = EDSData.copy()
```

```python
EDSData = EDSData.drop(columns=["Unnamed: 69"])
EDSData.head()
```

2°/ Valeurs manquantes

```python
EDSData.isnull().sum()
```

```python
pd.set_option("display.max_columns", None)
missing_percent = EDSData.isnull().mean()*100

pd.set_option("display.max_rows", None)
missing_percent.sort_values(ascending=False)
```

Objectif : supprimer les lignes qui sont toutes en NaN

```python
# Sélection des colonnes 4 à 67
cols = EDSData.columns[4:68]

# Supprimer lignes où toutes ces colonnes sont NaN
EDSData_clean = EDSData.dropna(subset=cols, how="all")
EDSData_clean.info()
```

```python
missing_percent_clean = EDSData_clean.isnull().mean()*100

pd.set_option("display.max_rows", None)
missing_percent_clean.sort_values(ascending=False)
```

Maintenant que les lignes full NaN sont supprimées j'observe que les annèes cette répartition :
* colonne quasiment vides (>85%) : 2017, 2016, 1971 à 1986
* Colonnes avec beaucoup de NaN mais encore acceptable (50-85%) : 2045 à 2085, 1970 à 1998, 2000 à 2014
* Colonne plutot complètes (<50%) : 2025, 2010
